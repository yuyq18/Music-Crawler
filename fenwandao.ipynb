```python
url = 'https://api.livelab.com.cn/performance/app/project/get_project_info?project_id=%s&v=1695113662390&retry=false'
```

```python
# black_list = ["大麦分销无座", "杜庞测试分销"]
start_id = 590
end_id = 798
# end_id = 589
```

```python
for i in range(start_id, end_id+1):
    print(f"Processing project {i}")
    project_url = url % str(i)
    r = requests.get(project_url, headers=headers)
    r_json = json.loads(r.text)
    new_detail_row = {}
    for col_ in details_df.columns:
        new_detail_row[col_] = r_json['data'][col_]
    details_df = pd.concat([details_df, pd.DataFrame([new_detail_row])], ignore_index=True)
    sleep(random.uniform(0.5, 2.5))
```

```python
# 过滤 测试项目 以及 类别不正确的项目
exist_files = os.listdir('./fenwandao/pre-docs')
exist_ids = [int(file.split('.')[0]) for file in exist_files]
print(len(exist_ids))
```

```python
problems_ids = []
```

```python
for i, exist_id in enumerate(exist_ids):
    with open(f'fenwandao/pre-docs/{exist_id}.json', 'r') as f:
        detail = json.load(f)
    if "测试" in detail['projectName'] or \
    '售卖' in detail['projectName'] or \
    '荧光棒' in detail['projectName'] or \
    'T恤' in detail['projectName'] or \
    '联名' in detail['projectName'] or \
    '周边' in detail['projectName'] or \
    '作废' in detail['projectName'] or \
    '预览' in detail['projectName']:
        problems_ids.append(exist_id)
        print(f"测试 or 商品 in {exist_id}")
    elif detail['subClassifyName'] == "未定义类型":
        problems_ids.append(exist_id)
        print(f"未定义类型 {exist_id}")
    else:
        with open(f'fenwandao/clean-docs/{exist_id}.json', 'w') as f:
            json.dump(detail, f, indent=4, ensure_ascii=False)
```
